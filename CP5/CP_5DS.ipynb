##Importações

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
index = pd.read_csv("salaries.csv")
index
```

```python
index.sample(5)
```

##**Evolução dos salários em empresas de pequeno, médio e grande porte**
##Gráfico de linhas


```python
small = index[index["company_size"] == "S"]
medium = index[index["company_size"] == "M"]
large = index[index["company_size"] == "L"]


salary1 = small.groupby('work_year')["salary_in_usd"].mean().reset_index()
salary2 = medium.groupby('work_year')["salary_in_usd"].mean().reset_index()
salary3 = large.groupby('work_year')["salary_in_usd"].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(salary1['work_year'], salary1['salary_in_usd'], label='Empresa pequena')
plt.plot(salary2['work_year'], salary2['salary_in_usd'], label='Empresa média')
plt.plot(salary3['work_year'], salary3['salary_in_usd'], label='Empresa grande')

plt.xlabel('Ano (dividido por semestre)')
plt.ylabel('Salário em dólares')
plt.title('Comparação de salários por ano')
plt.legend()
plt.show()
```

##**Salario médio de engenheiros de dados por ano**
##Gráfico de barras

```python
de = index[index["job_title"] == "Data Engineer"]

slary = de.groupby('work_year')["salary_in_usd"].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))

plt.bar(slary['work_year'], slary['salary_in_usd'], color='blue')

plt.xlabel('Ano')
plt.ylabel('Salário em dólares')
plt.title('Evolução do salário médio de engenheiros de dados por ano')

plt.show()
```

##**Evolução salárial no Canada**
##histograma

```python
cad = index[index["company_location"] == "CA"]

slario = cad.groupby('work_year')["salary_in_usd"].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))

plt.hist(slario['salary_in_usd'], bins=5, color='blue')

plt.xlabel('Salário em dólares')
plt.ylabel('Quantidade de pessoas')
plt.title('Média salárial no Canada')

plt.show()
```

##**Empregos nos EUA em 2024**
##Gráfico de pizza

```python
us = index[index["company_location"] == "US"]
us2024 = us[us["work_year"] == 2024]

jobs = us2024["job_title"].value_counts()
```

```python
plt.figure(figsize=(10, 10))

plt.pie(jobs, labels=jobs.index, autopct="%1.1f%%")
plt.title("Distribuição de empregos nos EUA em 2024")

plt.show()
```

##**Relação entre salário e ano**
##Gráfico scatter

```python
plt.figure(figsize=(10, 10))
plt.scatter( index["work_year"], index["salary_in_usd"], color="blue")

plt.xlabel("Ano")
plt.ylabel("Salário em dólares")
plt.title("Relação entre salário e ano")

plt.show()
```
